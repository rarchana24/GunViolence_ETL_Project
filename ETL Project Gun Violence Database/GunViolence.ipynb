```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
#Store Accidental Death CSV into DataFrame
Accidental_Death_Data = "Accidental_Death_Data.csv"
Accidental_Death_Data_df = pd.read_csv(Accidental_Death_Data)
Accidental_Death_Data_df.head()
```

```python
#Clean Accidental Death DataFrame - This is by removing the columns address and operations
new_accidental_death_df = Accidental_Death_Data_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()
new_accidental_death_df.head()
```

```python
#Rename Columns - rename the columns # killed to Num_killed, #injured to NumInjured
Accidental_Death_Data_df = new_accidental_death_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"US_State", "City Or County":"CityorCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
Accidental_Death_Data_df.head()
```

```python
#Store Accidental Death CSV into DataFrame
Accidental_Injuries = "Accidental_Injuries.csv"
Accidental_Injuries_Data_df = pd.read_csv(Accidental_Injuries)
Accidental_Injuries_Data_df.head()
```

```python
#Clean Accidental Injuries DataFrame - This is by removing the columns address and operations
new_accidental_injuries_df = Accidental_Injuries_Data_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()
new_accidental_injuries_df.head()
```

```python
#Rename Columns - rename the columns # killed to Num_killed, #injured to NumInjured
Accidental_Injuries_Data_df = new_accidental_injuries_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"US_State", "City Or County":"CityorCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
Accidental_Injuries_Data_df.head()
```

```python
#Store Mass Shootings CSV into DataFrame
Mass_Shootings_Data = "Mass_Shootings.csv"
Mass_Shootings_Data_df = pd.read_csv(Mass_Shootings_Data)
Mass_Shootings_Data_df.head()
```

```python
#Clean Mass shootings DataFrame - This is by removing the columns address and operations
new_mass_shootings_df = Mass_Shootings_Data_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()
new_mass_shootings_df.head()
```

```python
#Rename Columns - rename the columns # killed to Num_killed, #injured to NumInjured
Mass_Shootings_Data_df = new_mass_shootings_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"US_State", "City Or County":"CityorCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
Mass_Shootings_Data_df.head()
```

```python
#Merging the data (Mass shooting, accidental injuries)
combine_massshooting_accidentalinjury = pd.merge(Mass_Shootings_Data_df, Accidental_Injuries_Data_df, on="Incident_ID")
combine_massshooting_accidentalinjury
```

```python
# Drop the unnecessary columns
drop_columns_massshooting_accidentalinjury =combine_massshooting_accidentalinjury.drop(['Incident_Date_y', 'US_State_y', 'CityorCounty_y','Num_Killed_y','Num_Injured_y'], axis=1)
drop_columns_massshooting_accidentalinjury
Final_columns_massshooting_accidentalinjury = drop_columns_massshooting_accidentalinjury.rename(columns={
    "Incident ID":"Incident_ID", "Incident_Date_x":"Incident_Date", 
    "US_State_x":"US_State", "CityorCounty_x":"CityorCounty", 
    "Num_Killed_x":"Num_Killed", "Num_Injured_x":"Num_Injured"})

Final_columns_massshooting_accidentalinjury
```

```python
# Merging the data (Mass shooting,accidental death)
combine_massshooting_accidentaldeath = pd.merge(Mass_Shootings_Data_df, Accidental_Death_Data_df, on="Incident_ID")
combine_massshooting_accidentaldeath
```

```python
#Clean the data
# Drop the unnecessary columns
drop_columns_massshooting_accidentaldeath =combine_massshooting_accidentaldeath.drop(['Incident_Date_y', 'US_State_y', 'CityorCounty_y','Num_Killed_y','Num_Injured_y'], axis=1)
drop_columns_massshooting_accidentaldeath
Final_columns_massshooting_accidentaldeath = drop_columns_massshooting_accidentaldeath.rename(columns={
    "Incident ID":"Incident_ID", "Incident_Date_x":"Incident_Date", 
    "US_State_x":"US_State", "CityorCounty_x":"CityorCounty", 
    "Num_Killed_x":"Num_Killed", "Num_Injured_x":"Num_Injured"})

Final_columns_massshooting_accidentaldeath
```

```python
#Which States have the highest number of incidents (killed v. injured)
#Remove unecessary columns
MassShootingByState = Mass_Shootings_Data_df.drop(["Incident_ID","Num_Injured", "CityorCounty", "Incident_Date"], axis=1)
MassShootingByState

```

```python
#Mass Shootings Killed
MassShottingByState_sort_Killed = MassShootingByState.groupby(["US_State"]).sum().sort_values(["Num_Killed"], ascending=False)
MassShottingByState_sort_Killed

```

```python
#Mass Shootings Injured
#Which States have the highest number of incidents (killed v. injured)
#Remove unecessary columns
MassShootingByState_Injured = Mass_Shootings_Data_df.drop(["Incident_ID","Num_Killed", "CityorCounty", "Incident_Date"], axis=1)
MassShootingByState_Injured
```

```python
#Sort by State MassShootting Injured
MassShootingByState_sort_Injured = MassShootingByState_Injured.groupby(["US_State"]).sum().sort_values(["Num_Injured"], ascending=False)
MassShootingByState_sort_Injured
```

```python
#Remove unecessary columns Accidental injury killed
AccidentalInjury_Killed = Accidental_Injuries_Data_df.drop(["Incident_ID", "Incident_Date", "CityorCounty", "Num_Injured"], axis=1)
AccidentalInjury_Killed

AccidentalInjury_Injured = Accidental_Injuries_Data_df.drop(["Incident_ID", "Incident_Date", "CityorCounty", "Num_Killed"], axis=1)
AccidentalInjury_Injured

```

```python
#Accidental Injury Killed
Accidental_Injury_Killed_State_sort = AccidentalInjury_Killed.groupby("US_State").count().sort_values(["Num_Killed"], ascending=False)
Accidental_Injury_Killed_State_sort
```

```python
#Remove unecessary columns Accidental injury Injured
AccidentalInjury_Injured = Accidental_Injuries_Data_df.drop(["Incident_ID", "Incident_Date", "CityorCounty", "Num_Killed"], axis=1)
AccidentalInjury_Injured
#Accidental Injury Injured
Accidental_Injury_Injured_State_sort = AccidentalInjury_Injured.groupby("US_State").count().sort_values(["Num_Injured"], ascending=False)
Accidental_Injury_Injured_State_sort.reset_index()
```

```python
#Accidental Death Killed #Accidental_Death_Data_df

#Remove unecessary columns
Accidental_Death_Kill_State = Accidental_Death_Data_df.drop(["Incident_ID", "Num_Injured"], axis=1)
Accidental_Death_Kill_State

#Accidental Death Killed
Accidental_Death_Kill_State_sort = Accidental_Death_Kill_State.groupby("US_State").sum().sort_values(["Num_Killed"], ascending=False)
Accidental_Death_Kill_State_sort.reset_index()




```

```python
#Accidental death injured

#Remove unecessary columns
Accidental_Death_Injured_State = Accidental_Death_Data_df.drop(["Incident_ID", "Num_Killed"], axis=1)
Accidental_Death_Injured_State

#Accidental Death Killed
Accidental_Death_Injured_State_sort = Accidental_Death_Injured_State.groupby("US_State").sum().sort_values(["Num_Injured"], ascending=False)
Accidental_Death_Injured_State_sort.reset_index()
```

```python
mass_shooting_merge = pd.merge(MassShottingByState_sort_Killed, MassShootingByState_sort_Injured, on="US_State")
mass_shooting_merge.reset_index()
```

```python
accidental_injury_merge = pd.merge(Accidental_Injury_Killed_State_sort, Accidental_Injury_Injured_State_sort, on="US_State")
accidental_injury_merge.reset_index()
```

```python
accidental_death_merge = pd.merge(Accidental_Death_Kill_State_sort, Accidental_Death_Injured_State_sort, on="US_State")
accidental_death_merge.reset_index()
```
